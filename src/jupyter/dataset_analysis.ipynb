```python
%matplotlib notebook
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cases_samples = pd.read_csv('../../data/cases/cases_clean.csv')
cases_samples = cases_samples.drop(cases_samples.columns[0], axis = 1)
```

```python
cases_samples.head()
```

```python
# Percentage of none

number_of_values = cases_samples.shape[0] * cases_samples.shape[1]
number_of_nan = cases_samples.isnull().sum().sum()

print('Number of total genotypes: %5i' % number_of_values)
print('Percentage of Nan genotypes: %5.2f' % (number_of_nan/number_of_values) + '%')
```

```python
print('Number of samples: ' + str(len(cases_samples.index)))
```

```python
cases_samples.head()

plt.figure()
color_list = sns.color_palette(palette='hls', n_colors= 100)
for index in range(0, cases_samples.shape[0]):
    sns.distplot(cases_samples.iloc[[index]].astype(float).fillna(-2),\
                 kde=False,\
                 bins = 50, color = color_list[index])
    
plt.title('Histogram of genotypes')
plt.xlabel('Genotypes')
plt.xlim((0,15))
plt.show()
```

```python
spanish_samples = pd.read_csv('../../data/controls/spanish_clean.csv.gz', compression = 'gzip')
spanish_samples = spanish_samples.drop(spanish_samples.columns[0], axis = 1)
```

```python
print('Number of samples: ' + str(len(spanish_samples.index)))
```

```python
plt.figure()
color_list = sns.color_palette(palette='hls', n_colors= 1000)
for index in range(0, spanish_samples.shape[0]):
    sns.distplot(spanish_samples.iloc[[index]].astype(float).fillna(-2), \
                 bins = 100,\
                 color = color_list[index],\
                 kde=False)
plt.xlim((0,15))
plt.title('Histogram of genotypes')
plt.xlabel('Genotypes')
plt.show()
```
